```python
import datetime
import time

lis = []
for i in range(11,18):
    #Assign the required date/time
    date_time = datetime.datetime(2023, 10, i, 9, 00)
    print("unix_timestamp => ", (time.mktime(date_time.timetuple()))*1000)
    lis.append(int(time.mktime(date_time.timetuple()))*1000)

print(lis)
```

```python
import requests
import json

count = 11
for i in lis:
    # Setting up the API endpoint
    api_url = f"https://blockchain.info/blocks/{i}?format=json"

    try:
        response = requests.get(api_url)

        if response.status_code == 200:
            data = response.json()
            with open(f"blockchain_data_day_{count}.json",'w') as file:
                json.dump(data, file, indent = 4)
            print(f"Data written to blockchain_data_day_{count}.json")
        else:
            print(f"Failed to fetch data. Status code: {response.status_code}")
    except Exception as e:
        print(e)
    count+=1
```

```python
import os
import requests
import json

# Directory containing your JSON files
downloads_directory = r"/content"

# Create an empty list to store all the block data
all_block_data = []

# Function to fetch block information by hash
def fetch_block_info(hash_value):
    api_url = f"https://blockchain.info/rawblock/{hash_value}"
    try:
        response = requests.get(api_url)
        if response.status_code == 200:
            data = response.json()
            # Remove the "tx" key to exclude transaction data
            data.pop('tx', None)
            return data
        else:
            print(f"Failed to fetch data for hash {hash_value}. Status code: {response.status_code}")
    except Exception as e:
        print(f"Error fetching data for hash {hash_value}: {e}")
    return None

# Define the common part of the file name
common_file_name = "blockchain_data_day_"

# Iterate over the days
for day in range(11, 18):
    filename = f"{common_file_name}{day}.json"
    json_path = os.path.join(downloads_directory, filename)

    if os.path.exists(json_path):
        with open(json_path, "r") as file:
            block_data = json.load(file)

            # Extract the block hash from the list of block data and fetch additional info
            for block in block_data:
                hash_value = block.get("hash")
                if hash_value:
                    fetched_data = fetch_block_info(hash_value)
                    if fetched_data:
                        all_block_data.append(fetched_data)

# Save all the block data into a single output file
output_path = os.path.join(downloads_directory, "blockchain_data.json")
with open(output_path, 'w') as output_file:
    json.dump(all_block_data, output_file, indent=4)

print(f"Consolidated data written to {output_path}")
```
